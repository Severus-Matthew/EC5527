```python
#NAME - MANVI JHA
#ROLL - 1904014
# importing libraries
import math
import matplotlib.pyplot as plt
```

```python
#NAME - MANVI JHA
#ROLL - 1904014
#Plotting the fitness function Sin(x)/x [for –15 <= x <=15] Python code and output
a_x = []
out = []
for k in range(-150,150):
    a_x.append(k/10)
    if k == 0:
        out.append(1)
        continue
    n = k/10
    out.append((math.sin(n))/n)
plt.plot(a_x,out, color = 'blue', marker = "x") 
plt.title("GRADIENT DESCENT (Roll-1904014)") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.show()
```

```python
#NAME - MANVI JHA
#ROLL - 1904014
# Python code for executing the gradient descent algorithm and plot corresponding cost function
initial = -3
step_size = 2
x = 150 + initial*10
alpha = step_size * 10
ax = []
out = []
ar = []
out2 = []
cost = []
for k in range(-150,150):
    ax.append(k/10)
    if k == 0:
        out.append(1)
        continue
    n = k/10
    out.append((math.sin(n))/n)
p = 0
x1 = 0
while x1 != x:
    p = p+1
    ar.append(ax[x])
    grad = 10* (out[x] - out[x-1])
    x2 = round(x + grad * alpha)
    x1 = x
    x = x2
for k in range(p):
    l = ar[k]
    if l == 0:
        out2.append(1)
        continue
    out2.append((math.sin(l))/l)
for k in range(p):
    cost.append(1-out2[k])
print("Step Size",step_size)
print("Initial guess",initial)
print("Number of iteration required",p)
print("Maxima found at point",ar[p-1])
print("Maximum value",out2[p-1])
plt.subplot(2,1,1)
plt.plot(ax,out, color = 'red', marker = "o") 
plt.plot(ar,out2, color = 'green', marker = "o") 
plt.title("gradient decent") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.show()
plt.subplot(2,1,2)
plt.plot(cost, color = 'red', marker = "o") 
plt.title("cost function") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.show()
```
